### --- DESCRIPCIÓN DE CADA COLUMNA DEL DTAFRAME STUDEN_DATA ---

## 1.

**school**: Escuela a la que asiste el estudiante (GP: Gabriel Pereira, MS: Mousinho da Silveira).

**sex**: Sexo del estudiante (F: femenino, M: masculino).

**age**: Edad del estudiante (en años).

**address**: Tipo de domicilio (U: urbano, R: rural).

**famsize**: Tamaño de la familia (LE3: 3 o menos miembros, GT3: más de 3 miembros).

**Pstatus**: Estado civil de los padres (T: juntos, A: separados).

**Medu**: Nivel educativo de la madre (0: ninguno, 1: primaria, 2: 5º a 9º grado, 3: secundaria, 4: superior).

**Fedu**: Nivel educativo del padre (igual que Medu).

**Mjob**: Ocupación de la madre (teacher: docente, health: salud, services: servicios, at_home: en casa, other: otro).

**Fjob**: Ocupación del padre (igual que Mjob).

**reason**: Razón para elegir la escuela (home: cercanía, reputation: reputación, course: oferta de cursos, other: otro motivo).

**guardian**: Tutor principal del estudiante (mother: madre, father: padre, other: otro).

**traveltime**: Tiempo de viaje hasta la escuela (1: <15 min, 2: 15–30 min, 3: 30–60 min, 4: >1 hora).

**studytime**: Tiempo semanal dedicado al estudio (1: <2 horas, 2: 2–5 horas, 3: 5–10 horas, 4: >10 horas).

**failures**: Número de materias reprobadas anteriormente (0–3).

**schoolsup**: Apoyo educativo extra en la escuela (yes: sí, no: no).

**famsup**: Apoyo educativo extra familiar (yes: sí, no: no).

**paid**: Clases particulares de pago relacionadas con la materia (yes: sí, no: no).

**activities**: Participación en actividades extracurriculares (yes: sí, no: no).

**nursery**: Asistió a guardería infantil (yes: sí, no: no).

**higher**: Desea cursar estudios superiores (yes: sí, no: no).

**internet**: Acceso a internet en el hogar (yes: sí, no: no).

**romantic**: Tiene una relación amorosa (yes: sí, no: no).

**famrel**: Calidad de la relación familiar (1: muy mala, 5: excelente).

**freetime**: Tiempo libre después de la escuela (1: muy poco, 5: mucho).

**goout**: Frecuencia de salidas con amigos (1: muy poco, 5: mucho).

**Dalc**: Consumo de alcohol entre semana (1: muy bajo, 5: muy alto).

**Walc**: Consumo de alcohol en fines de semana (1: muy bajo, 5: muy alto).

**health**: Estado de salud actual (1: muy malo, 5: muy bueno).

**absences**: Número de ausencias escolares.

**final_grade**: Nota final del estudiante (G3, de 0 a 20).

### --- ANÁLISIS DE LA VARIABLE ABSENCES---
##### EJERCICIO 2.

```python
# Segmentamos el rendimiento en tres grupos según la nota final: Bajo: <=9, Medio: 10-14, Alto: 15-20
df['grupo_rendimiento'] = pd.cut(
    df['final_grade'],
    bins=[-1, 9, 14, 20],
    labels=['Bajo', 'Medio', 'Alto']
)
```

```python
# Analizamos las ausencias agrupando por grupo de rendimiento
ausencias_por_grupo = df.groupby('grupo_rendimiento')['absences'].describe()
```

```python
# Mostramos la tabla resumen (mean = media, std = desviación estandar)
print("Resumen estadístico de ausencias por grupo de rendimiento:")
display(ausencias_por_grupo)
```

A mayor rendimiento académico, menor es la cantidad promedio de ausencias escolares. Los estudiantes con bajo rendimiento faltan más a clase y muestran mayor variabilidad en sus ausencias.

```python
# Gráfico boxplot de ausencias por grupo de rendimiento
plt.figure(figsize=(8, 6))
sns.boxplot(x='grupo_rendimiento', y='absences', data=df, order=['Bajo', 'Medio', 'Alto'])
plt.title('Ausencias escolares según grupo de rendimiento')
plt.xlabel('Grupo de rendimiento')
plt.ylabel('Cantidad de ausencias')
plt.show()
```

El boxplot muestra que los estudiantes de rendimiento bajo tienden a tener más ausencias, con mayor dispersión y presencia de valores atípicos.

Los estudiantes de rendimiento medio presentan una cantidad de ausencias intermedia.

Los de rendimiento alto suelen tener menos ausencias y menor dispersión.

```python
# Gráfico de barras con la media de ausencias por grupo
plt.figure(figsize=(8, 4))
sns.barplot(x='grupo_rendimiento', y='absences', data=df, order=['Bajo', 'Medio', 'Alto'], estimator='mean')
plt.title('Media de ausencias por grupo de rendimiento')
plt.xlabel('Grupo de rendimiento')
plt.ylabel('Media de ausencias')
plt.show()
```

La media de ausencias disminuye a medida que aumenta el rendimiento académico.

#### Descripción del patrón:

Generalmente, se observa que a mayor tiempo de estudio semanal, la mediana y el promedio de la nota final tienden a aumentar. Esto sugiere una relación positiva entre dedicar más tiempo al estudio y obtener mejores resultados académicos.

### ---TRES VARIABLES CON MAYOR RELACIÓN AL BAJO DESEMPEÑO---
##### EJERCICIO 4.

```python
# Gráfico 1: Nota final según consumo de alcohol entre semana (Dalc)
plt.figure(figsize=(8, 6))
sns.boxplot(x='Dalc', y='final_grade', data=df)
plt.title('Nota final según consumo de alcohol entre semana')
plt.xlabel('Consumo de alcohol entre semana (1: muy bajo, 5: muy alto)')
plt.ylabel('Nota final')
plt.show()
```

#### Gráfico 1:
A medida que aumenta el consumo de alcohol durante los días de semana, la mediana de la nota final tiende a disminuir.
Los estudiantes con bajo consumo de alcohol suelen tener mejores notas, mientras que aquellos con consumo alto presentan notas finales más bajas y mayor dispersión.
Esto sugiere una relación negativa entre el consumo de alcohol entre semana y el rendimiento académico.

```python
# Gráfico 2: Nota final según tiempo de viaje (traveltime)
plt.figure(figsize=(8, 5))
sns.boxplot(x='traveltime', y='final_grade', data=df)
plt.title('Nota final según tiempo de viaje hasta la escuela')
plt.xlabel('Tiempo de viaje (1:<15min, 2:15-30min, 3:30-60min, 4:>1h)')
plt.ylabel('Nota final')
plt.show()
```

#### Gráfico 2:
El gráfico indica que, en general, los estudiantes con menor tiempo de viaje tienden a tener notas finales más altas. A medida que el tiempo de viaje aumenta, la mediana y el rango de las notas tienden a bajar, mostrando que los trayectos largos pueden estar asociados a un peor rendimiento académico.

```python
# Gráfico 3: Nota final según frecuencia de salidas con amigos (goout)
plt.figure(figsize=(8, 5))
sns.boxplot(x='goout', y='final_grade', data=df)
plt.title('Nota final según frecuencia de salidas con amigos')
plt.xlabel('Frecuencia de salidas con amigos (1: muy poco, 5: mucho)')
plt.ylabel('Nota final')
plt.show()
```

#### Gráfico 3:
Se observa que los estudiantes que salen muy poco o poco (valores bajos de goout) tienden a tener una mediana de nota final ligeramente más alta que aquellos que salen mucho.
A medida que aumenta la frecuencia de salidas, la mediana de la nota final tiende a disminuir levemente y la dispersión de las notas aumenta.
Esto sugiere que salir con amigos con mucha frecuencia podría estar asociado a un ligero descenso en el rendimiento académico, aunque la diferencia no es muy marcada.

```python
# Análisis conjunto: Nota final según consumo de alcohol entre semana y tiempo de estudio
plt.figure(figsize=(10, 6))
sns.boxplot(x='Dalc', y='final_grade', hue='studytime', data=df)
plt.title('Nota final según consumo de alcohol entre semana y tiempo de estudio')
plt.xlabel('Consumo de alcohol entre semana (1: muy bajo, 5: muy alto)')
plt.ylabel('Nota final')
plt.legend(title='Tiempo de estudio (1:<2h, 2:2-5h, 3:5-10h, 4:>10h)')
plt.show()
```

#### Análisis conjunto entre alcohol consumido entre semana y calificaciones

En general, los estudiantes que toman más alcohol entre semana suelen sacar peores notas.
Este efecto es más fuerte en quienes estudian poco.
En cambio, los que estudian más pueden compensar un poco el impacto negativo del alcohol, pero igual sus notas tienden a ser más bajas si consumen mucho.
Por lo tanto, estudiar ayuda, pero no elimina del todo el efecto negativo del alcohol en el rendimiento académico.

#### Justificación de elección:

Elegimos las variables de cantidad de alcohol ingerida entre semana, tiempo de viaje hasta la escuela y frecuencia de juntadas con amigos porque vimos que eran factores que podían afectar significativamente las notas de cada alumno y darían visualmente una relación clara y marcada entre problemas con alcohol, cansancio de viaje y pocas o demasiadas relaciones interpersonales.
